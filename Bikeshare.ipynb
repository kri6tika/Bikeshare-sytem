```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
cities = {'chicago':'Desktop/chicago.csv', 'new york':'Desktop/new_york_city.csv','washington':'Desktop/washington.csv'}
```

```python
print("Hey! Here you can explore bikeshare data of Chicago, new york and washington- available only in English")

def get_input():
    while True:
        c=input('\n Enter the city name').lower()
        if c in ['chicago', 'new york', 'washington']:
            break
        else:
            print('Invalid input.Please try again')
    return c

city=get_input()
file= cities[city]
df = pd.read_csv(file)
df.head()
```

```python
df.info()
```

```python
#Dropping the rows where at least one element is missing.
df_clean=df.dropna()
```

```python
df_clean.info()
```

```python
#Changing into format
df_clean['Start Time'] = pd.to_datetime(arg = df_clean['Start Time'])
df_clean['End Time'] = pd.to_datetime(arg = df_clean['End Time'])

#Adding new columns 
df_clean['month'] = df_clean['Start Time'].dt.month
df_clean['day_of_week'] = df_clean['Start Time'].dt.weekday_name
df_clean['hour'] = df_clean['Start Time'].dt.hour
df_clean.head()
```

```python
df_clean['Gender'] = df_clean['Gender'].map({'Female': 1, 'Male': 0})
df_clean['day_of_week'] = df_clean['day_of_week'].map({'Monday': 'mon', 'Tuesday': 'tue','Wednesday':'wed','Thursday':'thu','Friday':'fri','Saturday':'sat','Sunday':'sun'})
df_clean.head()
```

```python
# Calculating age and adding age column then removing column Birth year
df_clean['Birth Year']=df_clean['Birth Year'].astype(int)  
df_clean['c_year'] = int(datetime.now().year)
df_clean['age']=df_clean['c_year']-df_clean['Birth Year']
```

```python
df_clean.drop(columns=['Birth Year', 'c_year'])
df_clean.columns
```

```python
def get_filtered_by_month(df):
    m_input = input('Do you want a filter by month, y/n')
    if m_input =='y':
        while True:
            month = input("\n Enter month ( In first three letters like jan, feb...)").lower()
            if month in ['jan','feb','mar','apr','may','jun']:
                 break
            else:
                 print('Invalid_input.Please try again')
        df = df[df['month'] == month]
    return df
        
def get_filtered_by_day(df): 
    wd_input = input('Do you want a filter by day of the week, y/n')
    if wd_input =='y':
        while True :
            day = input("\n Enter month ( in first three letters like mon,tue... sun").lower()
            if day in ['mon','tue','wed','thu','fri','sat','sun']:
                break
            else:
                print('Invalid input.Please try again' )
        df = df[df['day_of_week'] == day]
    return df   
```

```python
#Filter
df_clean=get_filtered_by_month(df_clean)
df_clean=get_filtered_by_day(df_clean)
```
